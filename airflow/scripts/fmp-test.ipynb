```python
import requests
import json

CLIENT = "dc2478f367ba4d3"
SECRET = "h4eaqyvfnsrswpw"

def get_macro_events(start="2025-12-05", end="2025-12-19"):
    url = (
        "https://api.tradingeconomics.com/calendar"
        f"?d1={start}&d2={end}"
        f"&client={CLIENT}&client_secret={SECRET}"
    )

    r = requests.get(url)
    print("STATUS:", r.status_code)

    try:
        data = r.json()
    except Exception:
        print("❌ JSON decode failed:")
        print(r.text)
        return

    print("RAW RETURN:")
    print(data)

    # If API returns error
    if isinstance(data, dict) and ("error" in data or "message" in data):
        print("⚠️ API Error:", data)
        return

    # If wrapped inside list
    if isinstance(data, list):
        print(f"\n✔ Returned {len(data)} events\n")
        for ev in data[:10]:
            print(json.dumps(ev, indent=2))
        return

    print("⚠️ Unexpected format:", type(data), data)

get_macro_events()
```
